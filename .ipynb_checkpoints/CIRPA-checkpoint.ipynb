# DA-2 Task 2.10 — Key Findings Summary

## Transaction-Level EDA: Findings & Modelling Implications

---

### 1. Revenue per transaction is heavily right-skewed
**Finding:** Most transaction lines fall between £10–30, with a long tail of high-value outliers extending well beyond £100.

**So what:** Log transformation will be required before any distance-based ML model (K-Means). Raw monetary features will be dominated by extreme values and distort cluster centroids.

---

### 2. Strong seasonal revenue pattern in Q4
**Finding:** Revenue and transaction volume spike sharply in September–November, consistent with a gift retailer's pre-holiday season.

**So what:** Customers inactive in January–February may be seasonal buyers, not churners. The churn threshold in DA-6 must be set long enough (likely 90+ days) to avoid mislabelling seasonal purchase gaps as churn.

---

### 3. No Saturday transactions
**Finding:** Transaction volume drops to zero on Saturdays — the business does not operate on weekends (or at least Saturdays).

**So what:** The "most active day of week" feature will only span 6 values, not 7. This also confirms the retailer is likely B2B or wholesale-oriented rather than a typical consumer e-commerce site — consumer retail would show weekend activity.

---

### 4. Transactions peak between 10am–3pm
**Finding:** Near-zero activity before 7am and after 8pm, with volume concentrated during business hours.

**So what:** Reinforces the B2B/wholesale pattern. The "most active hour" feature will cluster heavily in the midday range, limiting its discriminative power for segmentation. Consider whether this feature adds value or just adds noise.

---

### 5. UK dominates at ~85% of revenue
**Finding:** UK accounts for roughly £9M of ~£10.6M total revenue and 20,000 of ~22,000 transactions.

**So what:** Filtering to UK-only for modelling loses only ~15% of revenue and is justified by concentration. However, this must be validated statistically in DA-7 with test results, not assumed from the charter.

---

### 6. International customers show 2× higher AOV and basket size
**Finding:** International AOV is ~£845 vs ~£447 for UK. International basket size is ~477 items vs ~224 for UK.

**So what:** International customers are likely wholesale buyers placing fewer but larger orders. The UK vs. international distinction is partly a proxy for retail vs. wholesale behaviour — important context when interpreting any geographic feature in modelling.

---

### 7. International return rate is 0% — likely a data artefact
**Finding:** UK return rate is ~28% while international shows exactly 0%.

**So what:** This is suspicious. Either returns for international orders are not recorded, or the negative-quantity convention differs. **Do not conclude that international customers never return products.** Investigate further in DA-7 and document this as a known data limitation.

---

### 8. April return rate spikes to ~13%
**Finding:** Monthly return rate peaks sharply in April (aggregated view), well above the ~4–8% baseline in other months.

**So what:** May coincide with post-holiday returns or a specific product quality issue. Return rate should be included as a customer-level feature in DA-3 because it varies meaningfully over time and likely across customers — it carries predictive signal for both churn and CLV.

---

### 9. Revenue per invoice shows bimodal pattern
**Finding:** A secondary bump appears around £250–350, suggesting two distinct populations: small retail orders and larger wholesale/bulk orders.

**So what:** Confirms the need for a wholesale/retail breakpoint in DA-5. If these two populations are mixed without acknowledgement, every average is misleading and clustering will produce muddled segments that blend fundamentally different customer types.

---

### 10. Customer-level revenue and quantity are heavily right-skewed
**Finding:** Long tails extend to £2,000+ revenue and 2,000+ units per customer, with most customers concentrated at low values.

**So what:** A small number of high-value customers disproportionately drive totals. This directly sets up the Pareto analysis in DA-4 — expect extreme concentration. Monetary and Frequency features will need log transformation before clustering.

---

### 11. Top 3 products dominate transaction volume
**Finding:** Products 85123A, 85099B, and 22423 each appear in 2,000+ invoices — roughly 2× the count of products ranked 10th–20th.

**So what:** Product concentration is high. "Product diversity" (unique StockCodes per customer) will be a meaningful feature — customers who buy only top products behave differently from those who explore the catalogue, and this likely correlates with purchase frequency and retention.

---

### 12. Specific products have structurally high return rates (12–25%)
**Finding:** Product 85017A leads at ~25% return rate, with several others consistently above 12%, even after filtering to products with 10+ transactions.

**So what:** High return rates on specific products suggest product quality or description accuracy issues, not random customer behaviour. In DA-7, cross-reference whether chronic returners are concentrated on these products — if so, the problem is product-driven rather than customer-driven, which changes the retention intervention strategy.

---

## Summary of Implications for Downstream Goals

| Finding | Impacts | Action |
|---|---|---|
| Right-skewed distributions | DA-3, ML-1 | Log-transform monetary and quantity features before clustering |
| Q4 seasonality | DA-6 | Set churn threshold ≥ 90 days to avoid seasonal false positives |
| B2B/wholesale signals | DA-5, ML-1 | Identify and handle wholesale vs. retail split before segmentation |
| UK dominance | DA-7 | Validate UK-only filter with statistical tests, not assumption |
| 0% international returns | DA-7 | Investigate and document as data limitation |
| April return spike | DA-3 | Include return rate as customer-level feature |
| Bimodal invoice revenue | DA-5 | Test for multimodality; define wholesale threshold |
| Product concentration | DA-3 | Engineer product diversity feature |
| High-return products | DA-7 | Cross-reference chronic returners with problem products |

# Feature Engineering

**Reference date**: 1/10/2011. All features computed relative to this date.

```python
df_positive = df[df['Quantity'] >= 0]
reference_date = pd.to_datetime('2011-10-01')

rfm_features = df_positive.groupby('CustomerID').agg(
    LastDay=('InvoiceDate', 'max'),
    FirstDay=('InvoiceDate', 'min'),
    Frequency=('InvoiceNo', 'nunique'),
    Monetary=('Revenue', 'sum'),
    Quantity=('Quantity', 'sum'),
    ProductDiversity=('StockCode', 'nunique'),
    Country=('Country','unique')
)

rfm_features['Recency'] = (reference_date - rfm_features['LastDay']).dt.days
rfm_features['DaysSinceFirstPurchase'] = (reference_date - rfm_features['FirstDay']).dt.days
rfm_features['FirstPurchaseMonth'] = rfm_features['FirstDay'].dt.month
rfm_features['Lastdaytofirstday'] = (rfm_features['LastDay'] - rfm_features['FirstDay']).dt.days
rfm_features['AverageRevenuePerOrder'] = rfm_features['Monetary'] / rfm_features['Frequency']
rfm_features['AvgItemsPerOrder'] = rfm_features['Quantity'] / rfm_features['Frequency']
rfm_features['UniqueItemsPerOrder'] = rfm_features['ProductDiversity']/rfm_features['Frequency']
#rfm_features['MonetaryVolatility'] = rfm_features['AverageRevenuePerOrder'].std()
rfm_features['IsUK'] = rfm_features['Country'].apply(
    lambda x: 1 if 'United Kingdom' in x else 0
)


rfm_features
```

```python
def compute_ipi(group):
    dates = group['InvoiceDate'].sort_values().unique()
    if len(dates)<=3:
        return 0
    diff = np.diff(dates).astype('timedelta64[D]').astype(int)
    return diff.mean()
    
def compute_mv(group):
    invoice_totals = group.groupby('InvoiceNo')['Revenue'].sum()
    if len(invoice_totals) < 3:
        return 0
    return invoice_totals.std()
    
rfm_features['InterPurchaseInterval'] = df_positive.groupby('CustomerID').apply(compute_ipi)
rfm_features['MonetaryVolatility'] = df_positive.groupby('CustomerID').apply(compute_mv)
#rfm_features['AverageRevenuePerOrder'].std()
rfm_features
```

```python
invoice_date = df_positive.groupby('CustomerID').agg({'InvoiceDate':'unique'})
invoice_date['MostActiveDay'] = df_positive.groupby('CustomerID')['InvoiceDate'].apply(
    lambda x: x.dt.dayofweek.mode()[0]
)

invoice_date['MostActiveHour'] = df_positive.groupby('CustomerID')['InvoiceDate'].apply(
    lambda x: x.dt.hour.mode()[0]
)

invoice_date['MostActiveMonth'] = df_positive.groupby('CustomerID')['InvoiceDate'].apply(
    lambda x: x.dt.month.mode()[0]
)
#invoice_date
features = rfm_features.join(invoice_date, on='CustomerID', how='left')
features.isna().sum()
```

```python
print(features.dtypes)

numeric = [
    'Recency',
    'DaysSinceFirstPurchase',
    'Lastdaytofirstday',
    'AverageRevenuePerOrder',
    'AvgItemsPerOrder',
    'UniqueItemsPerOrder',
    'MonetaryVolatility',
    'InterPurchaseInterval',
    'MostActiveDay',
    'MostActiveHour',
    'MostActiveMonth'
]
category = ['FirstPurchaseMonth', 'IsUK']

def distribution(df, feature_list):
    stats = df[feature_list].agg(['mean', 'median', 'skew', 'kurtosis']).T
    stats['high_skew'] = stats['skew'].abs() > 2
    print(stats.to_string())
    print(f"\nFeatures with |skew| > 2: {stats[stats['high_skew']].index.tolist()}")

    n = len(feature_list)
    cols = 3
    rows = (n + cols - 1) // cols
    fig, axes = plt.subplots(rows, cols, figsize=(15, 4 * rows))
    axes = axes.flatten()

    for i, feat in enumerate(feature_list):
        axes[i].hist(df[feat].dropna(), bins=30, color='#2E75B6', edgecolor='white')
        axes[i].set_title(feat)
        skew_val = df[feat].skew()
        axes[i].annotate(f'skew: {skew_val:.2f}', xy=(0.7, 0.85), xycoords='axes fraction', fontsize=9)

    for i in range(n, len(axes)):
        axes[i].set_visible(False)

    plt.tight_layout()
    plt.show()

distribution(features, numeric)
    
distribution(features, numeric)
```

AOV = NetRevenue / Frequency (average order value, net)
AvgItemsPerOrder = mean of Quantity sum per invoice
ProductDiversity = count of unique StockCodes purchased
InterPurchaseInterval = mean days between consecutive purchase dates (NaN or 0 for single-purchase customers — document your handling)
Tenure = last purchase date minus first purchase date (in days)
MonetaryVolatility = standard deviation of per-invoice revenue (high = unpredictable spender)
IntervalCV = standard deviation of inter-purchase intervals / mean of inter-purchase intervals (coefficient of variation — high = irregular buyer)
DaysSinceFirstPurchase = reference date minus first purchase date
AcquisitionMonth = month of first purchase (1–12)
MostActiveDay = mode of day-of-week across their purchases
MostActiveHour = mode of hour across their purchases
Create binary flag: 1 = UK, 0 = International


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# --- Correlation Heatmap ---
corr = features[numeric].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdBu_r', center=0,
            mask=np.triu(np.ones_like(corr, dtype=bool)),
            vmin=-1, vmax=1)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()

# --- Flag pairs with |r| > 0.80 ---
high_corr = []
for i in range(len(corr.columns)):
    for j in range(i + 1, len(corr.columns)):
        if abs(corr.iloc[i, j]) > 0.80:
            high_corr.append({
                'Feature 1': corr.columns[i],
                'Feature 2': corr.columns[j],
                'Correlation': round(corr.iloc[i, j], 3)
            })

high_corr_df = pd.DataFrame(high_corr).sort_values('Correlation', key=abs, ascending=False)
print("=== Highly Correlated Pairs (|r| > 0.80) ===")
print(high_corr_df.to_string(index=False))

# --- VIF Analysis ---
feature_data = features[numeric].dropna().replace([np.inf, -np.inf], 0)

vif_data = pd.DataFrame({
    'Feature': numeric,
    'VIF': [variance_inflation_factor(feature_data.values, i) for i in range(len(numeric))]
}).sort_values('VIF', ascending=False)

vif_data['Flag'] = vif_data['VIF'].apply(lambda x: '⚠ HIGH' if x > 10 else 'OK')

print("\n=== Variance Inflation Factor ===")
print(vif_data.to_string(index=False))
print(f"\nFeatures with VIF > 10: {vif_data[vif_data['VIF'] > 10]['Feature'].tolist()}")
```

# Revenue Concentration & Customer Value

```python
sorted_rev = features['Monetary'].sort_values(ascending=False).reset_index(drop=True)
total_rev = sorted_rev.sum()
n_customers = len(sorted_rev)

# Cumulative revenue %
cum_rev = sorted_rev.cumsum() / total_rev * 100
cum_customers = np.arange(1, n_customers + 1) / n_customers * 100

# Pareto chart
plt.figure(figsize=(10, 6))
plt.plot(cum_customers, cum_rev, color='#2E75B6', linewidth=2)
plt.plot([0, 100], [0, 100], '--', color='grey', linewidth=1, label='Perfect Equality')

# Annotate breakpoints
for pct in [1, 5, 10, 20]:
    idx = int(n_customers * pct / 100)
    rev_share = cum_rev.iloc[idx - 1]
    plt.annotate(f'Top {pct}% → {rev_share:.1f}% of revenue',
                 xy=(pct, rev_share),
                 xytext=(pct + 10, rev_share - 5),
                 arrowprops=dict(arrowstyle='->', color='black'),
                 fontsize=9)
    plt.plot(pct, rev_share, 'o', color='#C00000', markersize=6)

plt.xlabel('% of Customers')
plt.ylabel('% of Total Revenue')
plt.title('Pareto Curve — Revenue Concentration')
plt.legend()
plt.tight_layout()
plt.show()

# Print breakpoints
for pct in [1, 5, 10, 20]:
    idx = int(n_customers * pct / 100)
    rev_share = cum_rev.iloc[idx - 1]
    print(f"Top {pct}% of customers ({idx} customers) generate {rev_share:.1f}% of revenue")

sorted_asc = features['Monetary'].sort_values().values
n = len(sorted_asc)
index = np.arange(1, n + 1)
gini = (2 * np.sum(index * sorted_asc) / (n * np.sum(sorted_asc))) - (n + 1) / n
print(f"\nGini Coefficient: {gini:.3f}")
print(f"Interpretation: {'Extreme' if gini > 0.8 else 'High' if gini > 0.6 else 'Moderate'} revenue concentration")

boundaries = {
    'Platinum': (95, 100),
    'Gold': (75, 95),
    'Silver': (50, 75),
    'Bronze': (0, 50)
}

features['Tier'] = pd.cut(
    features['Monetary'].rank(pct=True) * 100,
    bins=[0, 50, 75, 95, 100],
    labels=['Bronze', 'Silver', 'Gold', 'Platinum']
)

tier_summary = features.groupby('Tier', observed=True).agg(
    CustomerCount=('Monetary', 'count'),
    TotalRevenue=('Monetary', 'sum'),
    MeanRevenue=('Monetary', 'mean')
)
tier_summary['% of Total Revenue'] = (tier_summary['TotalRevenue'] / total_rev * 100).round(1)
tier_summary['Percentile Range'] = ['0–50th', '50th–75th', '75th–95th', '95th–100th']

print("\n=== Customer Value Tiers ===")
print(tier_summary.to_string())
```

```python
# === Task 4.3 — Revenue-at-Risk ===
sorted_customers = features.sort_values('Monetary', ascending=False)

risk_data = []
for n in [10, 25, 50, 100]:
    top_n = sorted_customers.head(n)
    rev_lost = top_n['Monetary'].sum()
    pct_lost = rev_lost / total_rev * 100
    risk_data.append({'Lose Top N': n, '£ Lost': f'£{rev_lost:,.0f}', '% of Total': f'{pct_lost:.1f}%'})

risk_df = pd.DataFrame(risk_data)
print("\n=== Revenue-at-Risk ===")
print(risk_df.to_string(index=False))
```

## Risk Narrative
Our revenue is highly concentrated, with a Gini coefficient of 0.716. Just 43 customers (top 1%) generate 31.8% of all revenue, and 217 Platinum-tier customers (top 5%) account for over half of total revenue at 50.5%. The average Platinum customer is worth £20,664 — 65× more than the average Bronze customer at £320. Losing just our top 10 customers would eliminate £1.54M (17.3% of total revenue), and losing the top 100 would wipe out £3.61M (40.6%). This extreme concentration means the business's financial health depends on retaining a very small number of high-value accounts — a dedicated retention programme targeting the top 200 customers is not a cost, it is insurance against a multi-million pound revenue collapse.
